## 2. Model building

```python
tea_model = """

data {
    //num of observations
    int k;
    int n[k];
    //num of successes
    int s[k];

    //num of parameter distributions
    vector<lower=0> [k] alpha;
    vector<lower=0> [k] beta;
    int priorChoice;
    
}

parameters{
    //probability of success
    vector<lower=0, upper=1>[k] rate;
}

model {
     for (i in 1:k) {
    // prior
    if (priorChoice) {
      rate[i] ~ beta(alpha[i], beta[i]);
    } else {
      rate[i] ~ normal(0.5, 0.25);
      target += -log(normal_cdf(1, 0.5, 0.25) - normal_cdf(0, 0.5, 0.25));
    }
	  // likelihood
	  s[i] ~ binomial(n[i], rate[i]);
	}

}

"""
```

```python
data = {
    "k": 3,
    "n": np.array([8, 2, 5]),
    "s": np.array([3, 0, 5]),
    "alpha": np.repeat(1, 3),
    "beta": np.repeat(1, 3),
    "priorChoice": 0    
}
```

```python
posterior = stan.build(tea_model, data= data)
```

```python
fit = posterior.sample(num_chains=4, num_samples=1000)
df = fit.to_frame()
```
