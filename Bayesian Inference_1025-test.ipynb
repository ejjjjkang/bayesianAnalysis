```python
schools_code = """

data {
  int<lower=0> J;         // number of schools
  array[J] real y;              // estimated treatment effects
  array[J] real<lower=0> sigma; // standard error of effect estimates
}
parameters {
  real mu;                // population treatment effect
  real<lower=0> tau;      // standard deviation in treatment effects
  vector[J] eta;          // unscaled deviation from mu by school
}
transformed parameters {
  vector[J] theta = mu + tau * eta;        // school treatment effects
}
model {
  target += normal_lpdf(eta | 0, 1);       // prior log-density
  target += normal_lpdf(y | theta, sigma); // log-likelihood
}
"""
```

A Stan model requires at least three blocks, for each of data, parameters, and the model. 
data that will be used for sampling
parameter block specifies the relevant parameters


```python
schools_data = {"J" : 8, 
                "y" : [28,  8, -3,  7, -1,  1, 18, 12],
                "sigma" : [15, 10, 16, 11,  9, 11, 10, 18]}
```

```python
#compile stan model

posterior = stan.build(schools_code, data=schools_data)
```

```python
fit = posterior.sample(num_chains=4, num_samples=1000)
eta = fit["eta"]  # array with shape (8, 4000)
df = fit.to_frame()  # pandas `DataFrame, requires pandas
```

```python
sample_code = """

data{
	// number of observations
	int<lower = 1> n;
	// number of successes
	int<lower = 0, upper = n> s;
    real<lower = 0> a;
    real<lower = 0> b;
}
parameters{
	// probability of success
	real<lower = 0, upper = 1> rate;
}
model{
	// prior
	rate ~ beta(a, b);
	// likelihood
	s ~ binomial(n, rate);
}

"""
```

```python
sample_data = {
"s": 15, "n": 20, "a": 5, "b": 100
}
```

```python
sample_posterior = stan.build(sample_code, data=sample_data)
```

```python
fit = sample_posterior.sample(num_chains=4, num_samples=1000)
df = fit.to_frame()  # pandas `DataFrame, requires pandas
```
