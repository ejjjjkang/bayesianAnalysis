```python
method = np.concatenate([df["BloodPressure"], df["BMI"]])
outcome = np.concatenate([df["Outcome"], df["Outcome"]])
```

```python
print(len(method))
```

```python
## hypothesis setting
### H1. BloodPressure leads to diabetes
### H0. BMI is more effective to diabetes

""""
normal distribution
""""
```

```python
stan_code = """
            data {
            //num of observations
              int<lower=0> N;
            //num of success
              vector[N] outcome;
              real method[N];
            }
            
            parameters {
              // prob of success 
              real mu_A;
              real mu_B;
              real<lower=0> sigma;
            }
            
            model {
              //prior 
              mu_A ~ normal(69,19); //bloodpressure
              mu_B ~ normal(32,8); 
              sigma ~ cauchy(0, 5);
              
              //likelihood
              outcome ~ normal(mu_A + to_vector(method) * (mu_B - mu_A), sigma);
              }
"""
```

```python
data = {
"N": 2*len(df), 
"outcome": outcome, 
"method": method,
}
```

```python
model = stan.build(stan_code, data= data)
```

```python
fit = model.sample(num_chains=4, num_samples=1000)
```

```python
df = fit.to_frame()
```
